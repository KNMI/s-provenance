## NetCDF Provenance Visualisation

### SVG output

```python
import prov
import io
import StringIO
from prov.model import ProvDocument, ProvBundle, ProvException, first, Literal
from prov.dot import prov_to_dot
from prov.tests.test_model import AllTestsBase
from prov.tests.utility import DocumentBaseTestCase



def provToSvg(xml,output_f):
     
    xml_doc = StringIO.StringIO()
    xml_doc.write(str(xml))
    xml_doc.seek(0, 0)
    #print xml_doc
    doc=ProvDocument.deserialize(xml_doc,format="xml")
    dot = prov_to_dot(doc)
    return dot.create(format=output_f)
    

    


    

prov_doc=open("../../resources/testWPS_PROV").read()

svg_content=provToSvg(prov_doc,"svg")

with open("../../resources/testWPS_PROV.svg","w+") as text_file:
    text_file.write(str(svg_content))


from IPython.core.display import display, HTML
display(HTML(str(svg_content)))

    
# visualse NetCDF provenance in SVG
```

## PNG output

```python
png_content=provToSvg(prov_doc,"png")

with open("testWPS_PROV.png","w+") as text_file:
    text_file.write(str(png_content))
    text_file.flush()


from IPython.core.display import display, HTML
display(HTML("<img src='testWPS_PROV.png'/>"))
```
